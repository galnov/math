```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (5,5))
plt.axhline(y=0, color='black')
plt.axvline(x=0, color='black')

x = np.arange(-20,20,1)
y1 = np.power(x,2)        # x^2
y2 = np.power(x,2) + 100  # x^2+1
y3 = np.power(x-5,2)      # (x-5)^2
y4 = np.power(x,2) + x*5  # x^2 + 5*x

plt.plot(x,y1,'b')
plt.plot(x,y2,'g')
plt.plot(x,y3,'y')
plt.plot(x,y4,'r')
plt.show()
```
